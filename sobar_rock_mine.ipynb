```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
sonar_data=pd.read_csv('E:\Applied AI related materials\Python notebooks\Sonar_data.csv', header=None)
```

```python
sonar_data.shape
```

```python
sonar_data.head()
```

```python
X=sonar_data.drop(columns=60, axis=1)
Y=sonar_data[60]
```

```python
print(X.shape)
print(Y.shape)
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=1)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
print(X_train)
print(Y_train)
print(X_test)
```

```python
model=LogisticRegression()
```

```python
model.fit(X_train,Y_train)
```

```python
X_train_prediction=model.predict(X_train)
print("Accuray of the model on train data is ", accuracy_score(X_train_prediction,Y_train))
```

```python
X_test_prediction=model.predict(X_test)
print("Accuray of the model on test data is ", accuracy_score(X_test_prediction,Y_test))
```

```python
print("The accuracy of the model in percentage:",accuracy_score(X_train_prediction,Y_train)*100 )
```
